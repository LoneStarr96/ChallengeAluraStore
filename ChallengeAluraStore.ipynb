### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Calculando o faturamento total para cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Exibindo os resultados
print("Faturamento total da Loja 1:", faturamento_loja1)
print("Faturamento total da Loja 2:", faturamento_loja2)
print("Faturamento total da Loja 3:", faturamento_loja3)
print("Faturamento total da Loja 4:", faturamento_loja4)

# Preparando os dados para o gráfico
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Plotando o gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamentos, color=['blue', 'green', 'red', 'purple'])
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento Total (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Vendas por Categoria

```python
# Função para calcular e exibir as categorias mais populares
def categorias_populares(df, nome_loja):
    if 'Categoria do Produto' in df.columns:
        grouped = df.groupby('Categoria do Produto').size().sort_values(ascending=False)
        print(f"Categorias mais populares na {nome_loja}:")
        print(grouped)
        print("\n")

        # Plotando o gráfico de pizza
        plt.figure(figsize=(8, 8))
        plt.pie(grouped, labels=grouped.index, autopct='%1.1f%%', startangle=140)
        plt.title(f'Distribuição de Vendas por Categoria na {nome_loja}')
        plt.axis('equal')
        plt.show()
    else:
        print(f"A coluna 'Categoria do Produto' não existe na {nome_loja}.")

# Calculando para cada loja
categorias_populares(loja, "Loja 1")
categorias_populares(loja2, "Loja 2")
categorias_populares(loja3, "Loja 3")
categorias_populares(loja4, "Loja 4")
```

# 3. Média de Avaliação das Lojas


```python
# Calculando a média das avaliações para cada loja
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média das avaliações da Loja 1: {media_loja1:.2f}")
print(f"Média das avaliações da Loja 2: {media_loja2:.2f}")
print(f"Média das avaliações da Loja 3: {media_loja3:.2f}")
print(f"Média das avaliações da Loja 4: {media_loja4:.2f}")

# Preparando os dados para o gráfico
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_loja1, media_loja2, media_loja3, media_loja4]

# Plotando gráfico de linha com marcadores
plt.figure(figsize=(9, 6))
plt.plot(lojas, medias, marker='o', linestyle='-', color='purple', linewidth=2, markersize=10)
plt.title('Média de Avaliações dos Clientes por Loja', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Média de Avaliação', fontsize=12)
plt.ylim(min(medias) - 0.2, max(medias) + 0.2)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Adicionando os valores exatos em cada ponto
for i, v in enumerate(medias):
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Função para identificar produtos mais e menos vendidos
def produtos_vendidos(df, nome_loja):
    if 'Produto' in df.columns:
        vendas = df.groupby('Produto').size().sort_values(ascending=False)
        print(f"Vendas por produto na {nome_loja}:")
        print(vendas.to_string())
        max_vendas = vendas.max()
        mais_vendidos = vendas[vendas == max_vendas].index.tolist()
        print("\nProdutos mais vendidos (com {0} vendas):".format(max_vendas))
        print(mais_vendidos)
        min_vendas = vendas.min()
        menos_vendidos = vendas[vendas == min_vendas].index.tolist()
        print("\nProdutos menos vendidos (com {0} vendas):".format(min_vendas))
        print(menos_vendidos)
        print("\n")
        # Plotando o gráfico de barras
        vendas.plot(kind='bar', figsize=(10, 6))
        plt.title(f'Vendas por Produto na {nome_loja}')
        plt.xlabel('Produto')
        plt.ylabel('Número de Vendas')
        plt.show()
    else:
        print(f"A coluna 'Produto' não existe na {nome_loja}.")

# Calculando para cada loja
produtos_vendidos(loja, "Loja 1")
produtos_vendidos(loja2, "Loja 2")
produtos_vendidos(loja3, "Loja 3")
produtos_vendidos(loja4, "Loja 4")
```

# 5. Frete Médio por Loja

```python
# Calculando o custo médio de frete para cada loja
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

# Exibindo os resultados com duas casas decimais
print(f"Custo médio de frete da Loja 1: {media_frete_loja1:.2f}")
print(f"Custo médio de frete da Loja 2: {media_frete_loja2:.2f}")
print(f"Custo médio de frete da Loja 3: {media_frete_loja3:.2f}")
print(f"Custo médio de frete da Loja 4: {media_frete_loja4:.2f}")

# Preparando os dados para o gráfico de linha com marcadores
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias_frete = [media_frete_loja1, media_frete_loja2, media_frete_loja3, media_frete_loja4]

# Plotando o gráfico de linha com marcadores
plt.figure(figsize=(9, 6))
plt.plot(lojas, medias_frete, marker='o', linestyle='-', color='teal', linewidth=2, markersize=10)
plt.title('Custo Médio de Frete por Loja', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Custo Médio de Frete (R$)', fontsize=12)

# Ajustando os limites do eixo y
plt.ylim(min(medias_frete) * 0.9, max(medias_frete) * 1.1)

plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Adicionando os valores exatos acima de cada marcador
for i, v in enumerate(medias_frete):
    plt.text(i, v + (max(medias_frete) * 0.005), f"{v:.2f}",
             ha='center', fontsize=10, fontweight='bold', color='darkblue')

plt.tight_layout()
plt.show()
```
